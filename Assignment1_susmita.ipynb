```python
###import required libraries
import pymongo
import pandas as pd
```

```python
#Connecting with mongodb and Getting database collection named "Sample" from database
client = pymongo.MongoClient('localhost', 27017)
database = client['Sample']
collection = database['Sample']
```

```python
### Importing whole data from data base
sample_database = pd.DataFrame(list(collection.find()))
```

```python
# Exporting data to csv
sampleCSV = "sample.csv"
sample_database.to_csv(sampleCSV)
```

```python
### Closing the client connection
client.close()
```
